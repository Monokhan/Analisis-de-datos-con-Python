Â¿QuÃ© es un grÃ¡fico de lÃ­neas y por quÃ© usarlo?

Un grÃ¡fico de lÃ­neas o grÃ¡fico de lÃ­neas es un tipo de grÃ¡fico que muestra informaciÃ³n como una serie de puntos de datos llamados "marcadores" conectados por segmentos de lÃ­nea recta. Es un tipo bÃ¡sico de grÃ¡fico comÃºn en muchos campos. Use el diagrama de lÃ­neas cuando tenga un conjunto de datos continuo. Estos son mÃ¡s adecuados para visualizaciones de datos basadas en tendencias durante un perÃ­odo de tiempo.

Comencemos con un caso de estudio:

En 2010, HaitÃ­ sufriÃ³ un catastrÃ³fico terremoto de magnitud 7,0. El terremoto causÃ³ una gran devastaciÃ³n y pÃ©rdida de vidas y alrededor de tres millones de personas se vieron afectadas por este desastre natural. Como parte del esfuerzo humanitario de CanadÃ¡, el Gobierno de CanadÃ¡ intensificÃ³ sus esfuerzos para aceptar refugiados de HaitÃ­. Podemos visualizar rÃ¡pidamente este esfuerzo usando un grÃ¡fico de lÃ­neas:

Pregunta: Dibuje un grÃ¡fico lineal de la inmigraciÃ³n desde HaitÃ­ usando df.plot().

Primero, extraeremos la serie de datos para HaitÃ­.

```python
Haiti = df_can.loc['Haiti', years]# pasando en los aÃ±os 1980 - 2013 para excluir la columna 'total'
Haiti.head()
```

```python
#A continuaciÃ³n, trazaremos un diagrama de lÃ­neas agregando .plot() al marco de datos de HaitÃ­.
Haiti.plot()
```

pandas llenÃ³ automÃ¡ticamente el eje x con los valores del Ã­ndice (aÃ±os) y el eje y con los valores de la columna (poblaciÃ³n). Sin embargo, observe cÃ³mo los aÃ±os no se muestran porque son de tipo cadena. Por lo tanto, cambiemos el tipo de los valores de Ã­ndice a enteros para el trazado.

Podemos notar claramente cÃ³mo el nÃºmero de inmigrantes de HaitÃ­ se disparÃ³ desde 2010 cuando CanadÃ¡ intensificÃ³ sus esfuerzos para aceptar refugiados de HaitÃ­. Anotemos este pico en la grÃ¡fica usando el mÃ©todo plt.text().

```python
Haiti.plot(kind='line')
# anotar el terremoto de 2010.
# sintaxis: plt.text(x, y, etiqueta)
plt.title('Inmigracion desde Haiti')
plt.ylabel('Numero de inmigrantes')
plt.xlabel('AÃ±os')

plt.text(2000,6000, '2010 terromoto')#vea la nota abajo
plt.show()
```

Nota rÃ¡pida sobre los valores x e y en plt.text(x, y, label):

  Dado que el eje x (aÃ±os) es de tipo 'entero', especificamos x como un aÃ±o. El eje y (nÃºmero de inmigrantes) es de tipo 'entero', por lo que solo podemos especificar el valor y = 6000.

     plt.text(2000, 6000, '2010 Terremoto') # aÃ±os almacenados como tipo int

Si los aÃ±os se almacenaran como tipo 'cadena', necesitarÃ­amos especificar x como la posiciÃ³n del Ã­ndice del aÃ±o. Por ejemplo, el Ã­ndice 20 es el aÃ±o 2000, ya que es el aÃ±o 20 con un aÃ±o base de 1980.

     plt.text(20, 6000, '2010 Terremoto') # aÃ±os almacenados como tipo int

El diagrama de lÃ­neas es una herramienta Ãºtil para mostrar varias variables dependientes contra una variable independiente. Sin embargo, se recomienda no mÃ¡s de 5-10 lÃ­neas en un solo grÃ¡fico; mÃ¡s que eso y se vuelve difÃ­cil de interpretar.

Pregunta: Compare la tendencia de los 5 principales paÃ­ses que mÃ¡s contribuyeron a la inmigraciÃ³n a CanadÃ¡.

```python
#La respuesta correcta es:
#Paso 1: Obtenga el conjunto de datos. Recuerde que creamos una columna Total que calcula la inmigraciÃ³n acumulada por paÃ­s.
#Ordenaremos en esta columna para obtener nuestros 5 paÃ­ses principales usando el mÃ©todo pandas sort_values().

   # inplace = True # paramemter saves the changes to the original df_can dataframe
    df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)

    # get the top 5 entries
    df_top5 = df_can.head(5)

    # transpose the dataframe
    df_top5 = df_top5[years].transpose()

    print(df_top5)


    #Step 2: Plot the dataframe. To make the plot more readeable, we will change the size using the `figsize` parameter.
    df_top5.index = df_top5.index.map(int) # let's change the index values of df_top5 to type integer for plotting
    df_top5.plot(kind='line', figsize=(14, 8)) # pass a tuple (x, y) size



    plt.title('Immigration Trend of Top 5 Countries')
    plt.ylabel('Number of Immigrants')
    plt.xlabel('Years')


    plt.show()
```
